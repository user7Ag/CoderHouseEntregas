```python
paises=list(df.groupby('Country').sum().sort_values(by='Total', ascending=False).index) #Nombre de los paises
```

```python
valores=list(df.groupby('Country').sum().sort_values(by='Total', ascending=False)['Total']) #Total
```

```python
plt.figure(figsize=(10,8)) #Tamaño del grafico
```

```python
sns.set_style('ticks') #Color definido para los valores
```

```python
sns.barplot(x=paises[:5],y= valores[:5])

```
